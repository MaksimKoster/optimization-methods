## 1. Настроить вывод графики на базе PyPlot

```julia
using PyPlot 
surf(rand(30,40))
```

```julia
using PyPlot
using3D()

n = 100
x = range(-10,stop=10,length=n)
y = range(-10,stop=10,length=n)

xgrid = repeat(x',n,1)
ygrid = repeat(y,1,n)

a = 2
b = 4
z = zeros(n,n)
z .= (xgrid .^ 4) ./ (a .^ 4) .+ (ygrid ./ b) .^ 4 - (xgrid .^ 2 .* ygrid .^ 2) ./ (a .* b)
fig = figure("pyplot_surfaceplot",figsize=(10,30))
ax = fig.add_subplot(3,1,1,projection="3d")
plot_surface(xgrid, ygrid, z, rstride=2,edgecolors="k", cstride=2, alpha=0.8, linewidth=0.25)
xlabel("X")
ylabel("Y")
PyPlot.title("Fig 1")
```

```julia
using PyPlot
using3D()

n = 100
x = range(-10,stop=10,length=n)
y = range(-10,stop=10,length=n)

xgrid = repeat(x',n,1)
ygrid = repeat(y,1,n)

a = 2
b = 4
z = zeros(n,n)
z .= (xgrid .^ 2) .+ (ygrid .^ 2)
fig = figure("pyplot_surfaceplot",figsize=(10,30))
ax = fig.add_subplot(3,1,1,projection="3d")
plot_surface(xgrid, ygrid, z, rstride=2,edgecolors="k", cstride=2, alpha=0.8, linewidth=0.25)
xlabel("X")
ylabel("Y")
PyPlot.title("Fig 2")
```

```julia
using PyPlot
using3D()

n = 100
x = range(-10,stop=10,length=n)
y = range(-10,stop=10,length=n)

xgrid = repeat(x',n,1)
ygrid = repeat(y,1,n)

a = 2
b = 4
z = zeros(n,n)
z .= 1 ./ (1 .+ (xgrid .^ 2)) .+ 1 ./ (1 .+ (ygrid .^ 2))
fig = figure("pyplot_surfaceplot",figsize=(10,30))
ax = fig.add_subplot(3,1,1,projection="3d")
plot_surface(xgrid, ygrid, z, rstride=2,edgecolors="k", cstride=2, alpha=0.8, linewidth=0.25)
xlabel("X")
ylabel("Y")
PyPlot.title("Fig 3")
```
