# Показать, что множество точек минимума функции f(x) = 1/(1+x2), X=R на множестве X пусто и найти m0=inf f(x), x принадлежит X

## График

```julia
using PyPlot

a = -10
b = 10

x = range(a, b, length=100)
y = 1 ./ (1 .+ (x .* x))

PyPlot.plot(x, y)
```

## Точная нижняя грань

```julia
x_lim_right = Inf
y_right = 1 / (1 + (x_lim_right * x_lim_right))

println("Val: ", x_lim_right, ", Exact lower bound right: ", y_right)

x_lim_left = -Inf
y_left = 1 / (1 + (x_lim_left * x_lim_left))

println("Val: ", x_lim_left, ", Exact lower bound left: ", y_left)
```
