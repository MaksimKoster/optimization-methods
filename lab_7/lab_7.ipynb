# Исследовать на экстремум функцию при заданном условии (7 вариант)

```julia
f(x, y) = x^2 + y^2
```

```julia
g(x) = x + 1
```

```julia
using PyPlot
using3D()

n = 100
x = range(-10,stop=10,length=n)
y = range(-10,stop=10,length=n)

xgrid = repeat(x',n,1)
ygrid = repeat(y,1,n)

z = zeros(n,n)
z .= f.(xgrid, ygrid)
fig = figure("pyplot_surfaceplot",figsize=(10,30))
ax = fig.add_subplot(3,1,1,projection="3d")
plot_surface(xgrid, ygrid, z, rstride=2,edgecolors="black", cstride=2, alpha=0.8, linewidth=0.25)

for i in 0:200
    ax.plot(x, g.(x), zs=i, zdir="z")
end

xlabel("X")
ylabel("Y")
PyPlot.title("Fig 1")
```
